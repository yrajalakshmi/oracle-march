```python
#15.Create a bar chart of transaction counts per shopping_mall.
```

```python
 #22.Count transactions made on weekends.

```

```python
df['dayofweek'] = df['invoice_date'].dt.dayofweek
weekend_transactions = df[df['dayofweek'].isin([5, 6])]
count_weekends = len(weekend_transactions)
print(f"Number of transactions on weekends: {count_weekends}")
```

```python
#23.Find which shopping mall had the highest sales in December.
```

```python
dec_sales = df[df['invoice_date'].dt.month == 12].groupby('shopping_mall')['total_price'].sum()
top_mall = dec_sales.idxmax()   
top_sales = dec_sales.max()     
print(f"The shopping mall with the highest sales in December is {top_mall} with total sales of {top_sales}.")
```

```python
#25.Create a pivot table with average price by category and gender.
```

```python
#26.Plot daily revenue trends.
```

```python
daily_revenue = df.groupby(df['invoice_date'].dt.date)['total_price'].sum()
plt.figure(figsize=(10, 5))
daily_revenue.plot(kind='line', marker='o')
plt.title('Daily Revenue Trend')
plt.xlabel('Date')
plt.ylabel('Total Revenue')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#27.Determine which day of the week has the most transactions.
```

```python
df['day_of_week'] = df['invoice_date'].dt.day_name()
day_counts = df['day_of_week'].value_counts()
most_transactions_day = day_counts.idxmax()
print(f"The day with the most transactions is: {most_transactions_day}")
```
